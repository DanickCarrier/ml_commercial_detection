```python
import pandas as pd
import numpy as np
import scipy as sc
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score,recall_score,f1_score,matthews_corrcoef
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_svmlight_file
from sklearn.feature_selection import mutual_info_classif
```

```python
ndtv = load_svmlight_file("/kaggle/input/tv-news-channel-commercial-detection-dataset/NDTV.txt")
cnn = load_svmlight_file("/kaggle/input/tv-news-channel-commercial-detection-dataset/CNN.txt")
bbc = load_svmlight_file("/kaggle/input/tv-news-channel-commercial-detection-dataset/BBC.txt")
cnnibn = load_svmlight_file("/kaggle/input/tv-news-channel-commercial-detection-dataset/CNNIBN.txt")
timesnow = load_svmlight_file("/kaggle/input/tv-news-channel-commercial-detection-dataset/TIMESNOW.txt")
```

```python
x, y = sc.sparse.vstack([ndtv[0], cnn[0], bbc[0], cnnibn[0], timesnow[0]]), np.concatenate([ndtv[1], cnn[1], bbc[1], cnnibn[1], timesnow[1]])
```

```python
x = np.array(sc.sparse.csr_matrix.toarray(x))
```

```python
imp=pd.DataFrame(mutual_info_classif(df,y),index=df.columns)
imp.columns=['Importance']
imp.sort_values(by='Importance',ascending=False)
```

```python
from math import log10
documents = []

for j in range(len(tf.iloc[0, :])):
  count = sum( i > 0 for i in tf.iloc[:, j])
  documents.append(count)
```

```python
documents
```

```python
number_of_documents = sum(tf.iloc[0:, ])
number_of_documents
idf = []
for i in documents:
  value = 0
  if i > 0:
    value = log10(number_of_documents / i)
  else:
    value = 0

  idf.append(value)
```

```python
idf
```

```python
tf_idf = tf.copy()
```

```python
tf_idf *= idf
```

```python
for idx, value in enumerate(idf):
    if value == 0:
        tf_idf.drop(tf_idf.columns[idx], inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(tf_idf, y, test_size=0.50, random_state=102)
```

```python
pd.DataFrame(y_train).value_counts()
```

```python
pd.DataFrame(y_test).value_counts()
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True))
report
```
